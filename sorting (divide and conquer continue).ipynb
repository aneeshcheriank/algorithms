## selection sort
- find the minimum element
- place it at the begining of the array

```python
def selection_sort(lst):
    n = len(lst)
    
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            
            # find the minimum element in the list
            if lst[j] < lst[min_index]:
                min_index = j
                
        # swap the elements
        temp = lst[i]
        lst[i] = lst[min_index]
        lst[min_index] = temp
        
    return lst
        
lst = [12, 4, 6, 1, 56, 78, 23]
selection_sort(lst)
```

- time complexity: O(n<sup>2</sup>)

## merge sort

```python
def merge_sort(lst):
    n = len(lst)
    if n == 1:
        return lst
    
    half = n//2
    A = merge_sort(lst[:half])
    B = merge_sort(lst[half:])
    
    result = merge(A, B)
    return result

def merge(A, B):
    print(f'A: {A}')
    print(f'B: {B}')
    lst = []
    while len(A) != 0 and len(B) != 0:
    
        if A[0] <= B[0]:
            lst.append(A[0])
            A.remove(A[0])
            
        else:
            lst.append(B[0])
            B.remove(B[0])
            
    if len(A) != 0:
        lst.extend(A)
    else:
        lst.extend(B)
        
    print(f'lst: {lst}')
    print('___________')
            
    return lst

merge_sort(lst)
    
```



- 2 recursive calls (a)
- splitting array to 2 (b)
- running time is n (linear) (d)


- total steps = log(n)
- time taken in each level is the same
    - each level we need to go through every element in a single pass
- time taken = n (since the merging happening in linear time)

## Non Comparision based sorting algorithms
- only works if the elements are integers
- we scan the array and count the occurance

```python
def count_sort(lst):
    n = len(lst)
    output = [0 for _ in range(n)]
    max_el = max(lst)
    count = [0 for i in range(max_el+1)]
    
    for i in lst:
        count[i] += 1
#     print(count)
        
    # generate the cumilative sum of count
    # where to place the element in the lst 
    # count[lst[i]]-1 is the last postion, -1 becuase of 0 index
    for i in range(1, max_el+1):
        count[i] += count[i-1]
#     print(count)
        
    i = n-1
    while i >= 0:
        output[count[lst[i]]-1] = lst[i]
#         print(output)
        count[lst[i]] -= 1
        i -= 1
#         print(count)
#         print('-----------')
        
    return output

lst = [1, 3, 4, 5, 6, 1, 1, 2, 2, 3, 6, 5, 5]
count_sort(lst)

# time complexiy is O(m+n)
```

## Quick sort
- comparision based
- runtime: O(nlog(n))
- efficient

```python
def quick_sort(lst,l, r):
    # l: left
    # r: right
    if l >= r:
        return
    m = partition(lst, l, r)
    # now the postion m is in its final postion
    # now we need sort the array before m and array after m
    quick_sort(lst, l, m-1)
    quick_sort(lst, m+1, r)
    
def partition(lst, l, r):
    x = lst[l]
    j = l
    for i in range(l+1, r):
        # if the element is less than the pivot
        # swap the element 
            # increase j
            # swap to the incresed jth postion
        if lst[i] <= x:
            j += 1
            # swap the elements
            temp = lst[j]
            lst[j] = lst[i]
            lst[i] = temp
    # swap x; lst[l] and lst[j]
    temp = lst[l]
    lst[l] = lst[j]
    lst[j] = temp
    
    return j

lst = [12, 4, 6, 1, 56, 78, 23]
quick_sort(lst, 0, 7)
lst

# partition(lst, 0, len(lst))
```
