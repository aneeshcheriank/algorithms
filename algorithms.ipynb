   given a list of numbers from a1, a2, --- an, find max(ai.aj)

    2. array of integers
- output:
    1. <b>the max of the pair wise product</b>

```python
def find_fast_max_product(n: int, lst:list[int]) -> int:
    maximum = max(lst)
    lst.remove(maximum)
    maximum_second = max(lst)

    return maximum*maximum_second


lst = [i for i in range(1, 11)]
find_fast_max_product(len(lst), lst)
```

## Stress test
- 2 functions to produce the same result
    - solution

- generate random inputs
    - number of digits: random number
    - list: random numbers as elements
    
- check the output generated by both the functions are same
    - print the ouput, if the ouputs are different
    - trace the reson for difference

```python
# stress test
def find_max_product(n: int, lst: list[int]) -> int:
    max_product = -1
    for i in range(n-1):
        for j in range(i + 1, n):
            if j < n:
                product = lst[i] * lst[j]
                if product > max_product:
                    max_product = product
  
    return max_product


def stress_test(max_count):
 
  count = 0
  # infinate loop till conter over run or get a wrong answer

  while(True):
    # generate a random number
    # maximum 11
    n = random.randint(2, 12)     
    # generate random array:
    lst = []
    for _ in range(n):
        lst.append(random.randint(1, 100000))
        
    print(f'n: {n}, array: {lst}')
    
    # check the result of both the functions are same
    max_result = find_max_product(n, lst)
    fast_max = find_fast_max_product(n, lst)
      
    
    # condition to break
    # 1. count over run
    # 2. result not matching
    if fast_max != max_result :
        print('Wrong answer')
        print(f'fast_max result: {fast_max}, max_result: {max_result}')
        break   
        
    elif count >= max_count:
        break    
    count += 1


# exporting the libs
import random

# constant declarations
max_count = 10

stress_test(max_count)
```

## Fibonacci numbers
- Fn -> 0:n=0, 1:n=1, Fn-1+Fn-2: n> 1


```python
import time

def fibonacci(n):
    if n == 0:
        return [0]
    elif n == 1:
        return [0, 1]
    else:
        lst = [0, 1]
        for i in range(2, n+1):
            lst.append(lst[-1]+lst[-2])
            
    return lst[-1]

st = time.time()
print(fibonacci(30))
print(f'time taken: {time.time()-st} sec')
```

```python
# using recusrions
def fibonacci_recurssion(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recurssion(n-1)+fibonacci_recurssion(n-2)

st = time.time()
print(fibonacci_recurssion(30))
print(f'time taken: {time.time()-st} sec')
```

## Gratest Common Divisor


```python
# efficient function
def gcd(a, b):
    min_no = min(a, b)
    max_no = max(a, b)
    
    while max_no%min_no != 0:
        temp = min_no
        min_no = max_no%min_no
        max_no = temp
        
    return min_no

# the intutive function
def gcd_common(a, b):
    no = 1
    for i in range(2, min(a, b)):
        if (a%i == 0) and (b%i == 0):
            no = i
    return no

# function to time a function
def time_it(function):
    st = time.time()
    print(function)
    print(f'time taken to run: {time.time()-st} sec')

time_it(gcd_common(3918848, 1653264))
time_it(gcd(3918848, 1653264))
```

## Waiting time for patients
- in a clinic at 9.00, 5 patients came with treatment time for each is known, we have to assign the treatment for each person so as to minimize the total weighting time for all the patients

```python
t = [1, 2, 3, 4, 5]
n = 5

def min_total_waiting_time(t, n):
    waiting_time = 0
    treated = [0 for i in range(n)]
    
    # iterate through the entire patients
    for i in range(n):
        t_min = 9999
        min_index = 0
        
        # to find the minimum waiting time in the list
        # not treated
        for j in range(n):
            if treated[j] == 0 and t[j] < t_min:
                t_min = t[j]
                min_index = j
                
        waiting_time += (n-i-1)*t_min
        treated[min_index] = 1
        
    return waiting_time

min_total_waiting_time(t, n)

#O(n2)
```

## Party problem
- Many childeren came to a celebration. Organize them into the minimum possible numer of groups such that the age of any two childer in the same group is differ by at most two years. we need to reduce the number of groups.

```python
age = [2, 3, 4, 5, 6, 7, 8, 9, 10, 12, 14]
gap = 2

def party_groups_sorted(age, gap):
    
    groups = []
    left = 0
    n = len(age)
    
    while left < n:
        l, r = age[left], age[left]+gap
        groups.append((l, r))
        
        left += 1
        while left < n and age[left] <= r:
            left += 1
        
    return groups
        
party_groups_sorted(age, gap)

# running time: O(n)
```

```python
import random
import time

ages = [random.randint(5, 12) for _ in range(10000)]
ages.sort()

st = time.time()
gp = party_groups_sorted(ages, gap)
et = time.time() - st

print(f'time taken to run: {et} secs')
print(gp)
```

## Fractional knpsack
- we have items with weights [w1, w2, w3---wn] and values [v1, v2, v3 ---vn] what is the maximum value of item that can fit into a knapsack of capacity W. (we can take fraction of items)

- solution
    - try to take the item with maximum value per weight

```python
weights = [4, 6, 7, 8]
values = [20, 36, 49, 56]
capacity = 200

def fractional_knapsack(weighs, values, capacity):
    n = len(weights)
    unit_values = [v/w for v, w in zip(values, weights)]
    looked = [0 for _ in range(n)]
    
    profit = 0
    
    # sack full or no item to take
    while capacity > 0 and sum(looked) != n:
        
        max_value = 0
        index = 0
        
        # find the max value item
        for i, uv in enumerate(unit_values):
            if uv > max_value and looked[i] == 0:
                
                max_value = uv
                index = i
                
        # flag the item that has taken       
        looked[index] = 1
        profit += min(capacity, weights[index])*max_value
        capacity -= weights[index]
        
    return profit
            
fractional_knapsack(weights, values, capacity) 
```
